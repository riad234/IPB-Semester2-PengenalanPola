# 3. List

```python
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
list4 = ["abc", 34, True, 40, "male"]
print(list1);

# a. Menggabungkan list
gabung = list1 + list2 + list3 + list4
print("\na. Menggabungkan List")
print(gabung)

# b. Mengambil nilai
print("\nb. Mengambil nilai dari List")
print(list1[0])   # elemen pertama
print(list1[1])   # elemen kedua
print(list1[-1])  # elemen terakhir

# c. Menyisipkan nilai ke list
list2.insert(4, 11)
print("\nc. Menyisipkan nilai ke list2 (elemen 4 - 11)")
print(list2)

# d. Menghapus nilai dari list
print("\nd. Menghapus nilai dari list")
## Cara 1: remove() (berdasarkan nilai)
list1.remove("banana")
print(list1)
## Cara 2: pop() (berdasarkan index)
list1.pop(0)
print(list1)
## Cara 3: del
del list4[1]
print(list4)

# e. Mengurutkan isi list
# List harus berisi tipe data yang sama (semua angka atau semua string).
print("\ne. Mengurutkan isi list")
## Mengurutkan list angka
print("mengurutkan list angka")
list2.sort()
print(list2)
## Mengurutkan list string
print("mengurutkan list string")
list1.sort()
print(list1)
## Urutan terbalik
print("mengurutkan list secara terbalik")
list2.sort(reverse=True)
print(list2)
```
